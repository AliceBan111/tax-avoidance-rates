```python
import pickle
import pandas as pd
import re
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
df_test = pd.read_csv('test_fe.csv', index_col=0)
```

```python
df_test.columns = [re.sub(r'[^a-zA-Z0-9_]', '_', str(col)) for col in df_test.columns]
```

```python
X = ['etr_y_ma', 'etr_y_past', 'str', 'txt', 'str_cat__0_0875__0_192_', 'str_cat__0_28__inf_', 'cfc', 'revenue', 'pi', 'intant', 'intant_sqrt', 'diff', 'revenue_cat__0_174__1248_817_', 'WB_GDPpc', 'xrd_exists', 'revenue_cat__4233_587__inf_', 'ta', 'ta_log', 'ni', 'rd']
```

# Adaboost

```python
with open(r"..\\models\\adaboost_local_champion.pkl", "rb") as f:
    model_ada = pickle.load(f)

print(model_ada)
```

```python
y_pred_ada = model_ada.predict(df_test[X])
```

```python
y_true = df_test.etr.tolist()
rmse = np.sqrt(mean_squared_error(y_true, y_pred_ada))
mae = mean_absolute_error(y_true, y_pred_ada)
ada_model_score = pd.DataFrame([rmse, mae]).T.rename(columns ={0:"RMSE", 1:"MAE"})
```

```python
ada_model_score.rename(index={0: 'ada_model_score'},inplace=True)
ada_model_score
```

# XGBoostLSS

# Final comparison

```python
final_comparison = pd.concat([ada_model_score, cat_model_score, gbm_model_score, histgbm_model_score, lightgbm_model_score, xgboost_model_score, xgboostlss_model_score])
```

```python
final_comparison.sort_values("RMSE")
```

```python
final_comparison.rename({
    "ada_model_score": "AdaBoost",
    "cat_model_score": "CatBoost",
    "gbm_model_score": "GBM",
    "histgbm_model_score": "HistGBM",
    "lightgbm_model_score": "LightGBM",
    "xgboost_model_score": "XGBoost",
    "xgboostlss_model_score": "XGBoostLSS"
}, inplace=True)
```

```python
final_comparison = final_comparison.round(6)
final_comparison
```

```python
existing_models = pd.DataFrame({
    "RMSE": [0.167860, 0.085129, 0.115057, 0.082121, 0.094528, 0.084485],
    "MAE":  [0.091059, 0.061867, 0.073336, 0.058620, 0.067130, 0.061772]
}, index=["NAIVE MODEL", "OLS", "ARMA", "ARDL", "KNN", "SVR"])

final_comparison = pd.concat([final_comparison, existing_models])

final_comparison = final_comparison.sort_values(by="RMSE")
final_comparison
```

Based on the RMSE and MAE results, the tree-based models, especially HistGBM, XGBoostLSS, and XGBoost, achieve the best predictive performance, with the lowest errors. Traditional models like ARDL and SVR perform reasonably well but are slightly worse, while simple models such as ARMA and NAIVE MODEL have the highest errors and are less accurate. Overall, ensemble tree-based methods clearly outperform the others in this dataset.
